```python
from pathlib import Path
import pandas as pd
import requests
from bs4 import BeautifulSoup
from maple.nlp.crawler.auto import Automation
from selenium.common.exceptions import UnexpectedAlertPresentException
import re
import time
import random



fpath = Path('/Users/kyu/PycharmProjects/EzCommerce/data/prdNames2.csv')
prd_names = pd.read_csv(fpath).iloc[:, 0].to_list()
lowest_prices = []
lowest_hrefs = []

target_url = 'https://search.shopping.naver.com/search/all.nhn?'
sort_tp = 'price_asc'
# min_price = '0'
# max_price = '19,598'
```

```python
MAX_TRY = 5
TIME_WAIT = 2
```

```python
crawler = Automation()
for prd in prd_names:
#     query = f'{target_url}query={prd}&sort={sort_tp}&minPrice={min_price}&maxPrice={max_price}'
    url = f'{target_url}query={prd}&sort={sort_tp}'
    html = requests.get(url)
    html_doc = html.text
    soup = BeautifulSoup(html_doc, 'html.parser')
    results = soup.find_all('div', class_='info')
    
#     if len(results):
#         pass
#     else:
#         print(f'Given {prd} is at the lowest price!')
#         continue
    lprice = None
    i = 0
    max_try = min(MAX_TRY, len(results))
    while lprice is None:
        if i == max_try:
            print('==========================')
            print(f" CHECK OUT {prd} MANUALLY ")
            print('==========================')
            break
        print(f'\nFinding lowest price... trial {i}')
        item = results[i]
        
        lprice_href = item.find('a', class_='link').get('href')
        
        if lprice_href.startswith('http://cr2.shopping.naver.com/adcr.nhn'):
            # if goes to another compare page, just skip it
            i += 1
            continue
            
        lprice = re.sub(r'[^\d+]', '', item.find('span', class_='num _price_reload').text) # digits
        lprice_alt = '{:,}'.format(int(lprice))  # digits with marks every thousands
        check = re.compile(f'({lprice}|{lprice_alt})')
        
        print(f'\tFound a candidate lowest price: {lprice}')
#         print(lprice_href)
        
        try:
            crawler.driver.get(lprice_href)

            is_alert, msg = crawler.is_alert()
            if is_alert:
                print(f'\tAlert: {msg}')

            time.sleep(TIME_WAIT + random.random())
            lprice_html = crawler.driver.page_source
            found = check.search(lprice_html)
        
        except UnexpectedAlertPresentException:
            i += 1
            lprice = None
            print(f'\tCandidate Not Found')
            continue
        
        if found is None:
            print(f'\t{lprice} is verfied as ** FAKE **')
            lprice = None
        else:
            print(f'\t{lprice} is verifies as ** REAL **')
            
        i += 1
        
    if lprice is not None:
        print(f'{prd} --  lowest price: {lprice}')
        lowest_prices.append(lprice)
        lowest_hrefs.append(lprice_href)
    else:
    
        lowest_prices.append(None)
        lowest_hrefs.append(None)



assert len(prd_names) == len(lowest_prices) 
assert len(lowest_prices) == len(lowest_hrefs)
```
